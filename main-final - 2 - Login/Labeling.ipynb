Proses Pelabelan

```python
import pandas as pd
df = pd.read_csv('Hasil Praprocessing Data')

def read_lexicon(file_path):
    lexicon = {}
    with open(file_path, 'r') as file:
        for line in file:
            line = line.strip().split('\t')
            if len(line) == 2:
                word = line[0]
                weight = line[1]
                if weight.lower() == 'weight':
                    continue
                try:
                    weight = int(weight)
                except ValueError:
                    weight = 0
                lexicon[word] = weight  # <- dipindah keluar dari except
    return lexicon

def calculate_sentence_weight(row, negatif_lexicon, positif_lexicon):
    words = row['tokens_stemmed']  # <- perbaikan dari word → words
    total_weight = 0
    for word in words:
        negatif_weight = negatif_lexicon.get(word, 0)
        positif_weight = positif_lexicon.get(word, 0)
        total_weight += positif_weight + negatif_weight  # <- asumsi: positif menambah, negatif mengurangi
    return total_weight

file_path_negatif = 'negative.tsv'
file_path_positif = 'positive.tsv'

negatif_lexicon = read_lexicon(file_path_negatif)
positif_lexicon = read_lexicon(file_path_positif)

def determine_label(bobot):
    if bobot >= 0:
        return "positif"
    else:
        return "negatif"

# Hitung bobot dan label
df['bobot_nilai_total'] = df.apply(lambda row: calculate_sentence_weight(row, negatif_lexicon, positif_lexicon), axis=1)
df['label'] = df['bobot_nilai_total'].apply(determine_label)

# Tampilkan hasil
display(df[['tokens_stemmed', 'bobot_nilai_total', 'label']])
```
